```python
def makeAxis(title, tickangle):
    return {
      'title': title,
      'titlefont': { 'size': 14 },
      'tickangle': tickangle,
      'tickfont': { 'size': 12 },
      'tickcolor': 'rgba(0,0,0,0)',
      'ticklen': 5,
      'showline': True,
      'showgrid': True
    }

fig = go.Figure(go.Scatterternary({
    'mode': 'markers',
    'a': [i for i in df['Temperature'].values],
    'b': [i for i in df['Humidity'].values],
    'c': [i for i in df['Step count'].values],
     'text': [i for i in df['Stress Level'].values],
    'marker': {
        'symbol': 100,
        'color': df['Stress Level'],
        'size': 14,
        'line': { 'width': 2 }
    }
}))

fig.update_layout({
    'ternary': {
        'sum': 100,
        'aaxis': makeAxis('Temperature', 0),
        'baxis': makeAxis('<br>Humidity', 45),
        'caxis': makeAxis('<br>Step count', -45)
    },
    'annotations': [{
      'showarrow': False,
      'text': 'Temperature vs Humidity vs Step count',
        'x': 0.5,
        'y': 1.3,
        'font': { 'size': 20 }
    }]
})

fig.show()
```

```python
y = df["Stress Level"]
X = df.drop(columns=["Stress Level"])
print(X.shape, y.shape)
```

## Create KNN Classification model

```python
#  Split dataset into train & test components
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
print(X.shape, y.shape)
```

```python
# # Scale dataset
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# X_train = scaler.fit_transform(X_train)
# X_test = scaler.transform(X_test)

# len(X_test)
```

```python
# Use the KNN classifier to fit data:
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=10)
clf= classifier.fit(X_train, y_train) 
clf
```

```python
# Predict y data with classifier: 
y_predict = classifier.predict(X_test)
```

```python
# Print training results and accuracy score: 
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print(confusion_matrix(y_test, y_predict))
print("\nThe accuracy score is: ", accuracy_score(y_test, y_predict))
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# Print r2 and mse scores
mse = mean_squared_error(y_test, y_predict)
r2 = clf.score(X_test, y_test)
print(f"mse: {mse}")
print(f"r2: {r2}")
```

```python
# Print classfication report
print(classification_report(y_test, y_predict)) 
```

## Save the model using pickle

```python
# Create pickle file
knnPickle = open('../models/KNNmodel_stress_prediction', 'wb') 

# Save the trained model to file
pickle.dump(clf, knnPickle)      
knnPickle.close()
```
