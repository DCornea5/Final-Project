```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../data/heart_2020_cleaned.csv")
df
```

```python
# setting up dependent variable
y = df["Stroke"]
```

```python
# encoded target column heart disease
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
# encoded_y = [[x] for x in encoded_y]
encoded2d_y = np.reshape(encoded_y,(-1,1))
encoded2d_y

```

```python
data = df.drop(["Stroke"], axis = 1)
X= pd.get_dummies(data)
X.columns
```

```python
# remove unnecessary columns
X = X.drop(["Smoking_No","AlcoholDrinking_No","HeartDisease_No","DiffWalking_No","PhysicalActivity_No","Asthma_No","KidneyDisease_No","SkinCancer_No"],axis = 1)
```

```python
# Store column names
feature_names = X.columns
feature_names
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, encoded_y, random_state=42)
```

## Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=1000)
```

```python
classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# predicted = classifier.predict(X)
# mse = mean_squared_error(y, predicted)
r2 = classifier.score(X_test, y_test)
print(f"Logistic Regression score: %.04f" %r2)
```

## Decision tree model

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
r2 = clf.score(X_test, y_test)
print(f"Tree score: %.04f" %r2)
```

## Random forest model

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50) 
rf = rf.fit(X_train,y_train)
rf.score(X_test, y_test)
```

```python
# calculate feature importance
importances = rf.feature_importances_
# We can sort the features by their importance
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```
